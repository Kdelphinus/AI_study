# **1. One-hot Encoding이란?**
---

원-핫 인코딩은 선택해야 하는 선택지의 개수만큼의 차원을 가지면서, 각 선택지의 인덱스에 해당하는 원소에는 1, 나머지 원소는 0의 값을 가지도록 하는 표현 방법입니다. 

예를 들어 고양이, 강아지, 냉장고라는 3개의 선택지가 있다고 합시다. 원-핫 인코딩을 위해선 우선 각 선택지에 순차적으로 정수 인덱스를 부여합니다. 임의로 강아지 - 0, 고양이 - 1, 냉장고 - 2를 부여했다고 합시다. 이때 각 선택지에 대해 원-핫 인코딩이 된 벡터는 다음과 같습니다.

- 강아지 = [1, 0, 0]
- 고양이 = [0, 1, 0]
- 냉장고 = [0, 0, 1]

총 선택지는 3개였으므로 위 벡터들은 전부 3차원의 벡터가 되었습니다. 그리고 해당 선택지의 인덱스를 1, 나머지를 0의 값을 가집니다. 이와 같이 원-핫 인코딩으로 표현된 벡터를 **one-hot vector**라고 합니다.


# **2. 원-핫 벡터의 무작위성**
---

꼭 실제값을 원-핫 벡터로 표현해야만 다중 클래스 분류 문제를 풀 수 있는 것은 아니지만, 대부분의 다중 클래스 분류 문제가 각 클래스 간의 관계가 균등하기에 원-핫 벡터는 이러한 점을 표현할 수 있는 적절한 표현 방법이다. 

다수의 클래스 문제는 클래스의 개수만큼 레이블이 필요합니다. 이때 직관적으론 {강아지, 고양이, 호랑이}를 각각 0, 1, 2를 레이블로 가지는 정수 인코딩을 할 수 있지만 일반적인 다중 클래스 분류 문제는 정수 인코딩보다 원-핫 인코딩을 사용하는 것이 더 클래스의 성질을 잘 표현했다고 할 수 있습니다. 

예를 들어 평균 제곱 오차(MSE)를 사용할 때, 강아지와 고양이 사이의 오차보다 강아지와 호랑이 사이의 오차가 더 커지게 된다. 이는 기계에게 강아지와 고양이가 더 가깝다는 정보를 전해주는 것과 같습니다. 만약 3개가 아닌 더 많은 것들을 정수 인코딩한다면 그 오차는 굉장히 커질 것입니다.

이러한 정수 인코딩은 {1층, 2층, 3층, 4층}, {10대, 20대, 30대, 40대}와 같은 경우 성립하지만 일반적인 분류 문제에서는 순서의 의미가 없기에 성립하지 않습니다. 그렇기에 모든 클래스 간의 관계를 균등하게 분배하는 원-핫 인코딩을 사용합니다. 

$$((1, 0, 0) - (0, 1, 0))^2 = (1 - 0)^2 + (0 - 1)^2 + (0 - 0)^2 \\
((1, 0, 0) - (0, 0, 1))^2 = (1 - 0)^2 + (0 - 0)^2 + (0 - 1)^2$$

다른 말로 모든 클래스에 대해서 원-핫 인코딩을 통해 얻은 원-핫 벡터들은 모든 쌍에 대해 유클리드 거리를 구해도 모두 동일합니다. 이렇게 원-핫 벡터의 표현 방법이 무작위성을 가진다는 점을 표현할 수 있습니다.
